```
import numpy as np

def add(A, B):
  if A.shape != B.shape:
      raise RuntimeError("Matrices A and B don't have the same shape")
  C = np.zeros(A.shape)
  for i in np.arange(A.shape[0]):
    for j in np.arange(A.shape[1]):
        C[i][j] = A[i][j] + B[i][j]
  
  return C

A = np.array([[0, 4], [7, 0], [3, 1]])
B = np.array([[1, 2], [2, 3], [0, 4]])

C = add(A, B)
print(C)
```

```
def subtract(A, B):
    if A.shape != B.shape:
      raise RuntimeError("The matrices A&B are not of the same shape")
    
    nrows = A.shape[0]
    ncols = A.shape[1]
    # Define a new matrix which stores the sum
    C = np.zeros(A.shape)
    
    # Iterate through rows
    for i in np.arange(nrows):
        # Iterate through cols
        for j in np.arange(ncols):
            C[i][j] = A[i][j] - B[i][j]
    
    return C
```

```
B = np.array([[1, 6], [9, 3]])
I = np.eye(2)
C = subtract(B, I)
print(C)
```
