```julia
t = 2
# Position of particle
def position(t):
  return np.array([2.0 * t**2, (2.0 + 3.0*t), 5 * t])

# Velocity of particle
def velocity(t):
  return np.array([4.0 * t, 3.0, 5.0])

## Velocity at t = 2
v_t2 = velocity(2)
print(v_t2)

## Speed
print(np.linalg.norm(v_t2))

## Average velocity
t1 = 3
t2 = 1
v_avg = (position(t1) - position(t2))/(t1 - t2)
print(v_avg, np.linalg.norm(v_avg))
```
